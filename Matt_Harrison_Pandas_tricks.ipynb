```python
# can go deeper and apply multiple aggregates

def second_to_last(ser):
    return ser.iloc[-2]
```

```python
(autos2
 .assign(country=autos2.make.apply(country))
 .groupby(['year', 'country'])
 .agg(['mean', min, second_to_last])
)
```

```python
def vals_gt(df_, num):
    return df_[df_.gt(num)].dropna()

(autos2
 .assign(country=autos2.make.apply(country))
 .groupby(['year', 'country'])
 .count()
 .pipe(vals_gt, 700)
)
```
